```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import time

# initiate selenium web driver for chrome
driver = webdriver.Chrome()

# navigate to the ESPN Fantasy Football home webpage
driver.get('https://fantasy.espn.com/fantasy/football/')

try: 

    # click on the "Home" button, even though the text is "Log In"
    home_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Log In")))
    print("Home button is clickable")

    home_button.click()

    # click on the actual "Log In" button
    login_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a[data-affiliatename='espn'][data-language='en']")))
    print("'Log in' button is clickable")

    login_button.click()

    try: 

        # ensure the driver is working in the disney login page frame
        WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.XPATH,"//iframe[@name='oneid-iframe']")))
        print("Frame switched")

        try:

            # ensure the username input field is visible on screen
            username = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH,"//input[@class='input-InputIdentityFlowValue form-control']" )))
            print("Username input field is visible")

            # have to select the input field before we can send keys
            username.click()

            # enter username
            username.send_keys("lhdflemming@gmail.com")

            # ensure the "Continue" button is clickable
            continue_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,"//button[@class='btn btn-submit btn-primary']")))
            print("Continue button is clickable")

            continue_button.click()

            try: 

                # ensure the password input field is visible on screen
                password = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH,"//input[@class='input-InputPassword form-control']")))
                print('Password input field is visible')

                # have to select the input field before we can send keys
                password.click()

                # enter password
                password.send_keys("Jebronlames1687!")

                # ensure "Login" button is clickable and then click it
                login_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[@class='btn btn-submit btn-primary']")))

                login_button.click()
            

                try:

                    # ensures that the sign in button is available
                    sign_in = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Sign In")))
                    print("Ready to select league")

                    league_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[@name='&lpos=fantasy:fantasy:ffl:5:team']")))
                    print("League selection successful")
                    league_link.click()
                    
                    # ensure the add player button is available
                    players_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//li[@class='players NavSecondary__Item']/a[@class='AnchorLink NavSecondary__Link hasSub']")))
                    players_button.click()

                    # go to player search bar
                    search_bar = WebDriverWait(driver, 10).until(
                        EC.visibility_of_element_located((By.XPATH, "//input[@class='form__control']"))
                    )
                    search_bar.click()

                    # enter player to add/drop into search bar
                    search_bar.send_keys("Brock Bowers")

                    #time.sleep(2)

                    try:

                        # lazy and just selecting the first result because I'm using player's full name
                        # I could get the player ID and select the player from the dropdown using that
                        player_to_be_selected = WebDriverWait(driver, 10).until(
                            EC.element_to_be_clickable((By.XPATH, "//button[@data-idx='0']"))
                        )

                        player_to_be_selected.click()

                    except:
                        print("Player could not be selected")
                
                except:
                    print("League link failed")

            except:
                print('Password failed')

        except:
            print("Input failed")

    except:
        print("Frame didn't switch")

except TimeoutException:
    print("Fail")
```
